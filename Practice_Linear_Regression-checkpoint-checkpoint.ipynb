```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
A=pd.read_csv(r"C:\Users\hp\Desktop\Datasets\temperatures.csv")
```

```python
A
```

```python
A.dropna()
```

```python
A.describe()
```

```python
x=A[["YEAR"]]
y=A[["JAN"]]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
model=reg.fit(xtrain,ytrain)
```

```python
plt.title("Temprature Predictions")
plt.xlabel("Year")
plt.ylabel("Jan")
plt.scatter(xtrain,ytrain,color="red")
plt.plot(xtrain,reg.predict(xtrain),color="blue")
```

```python
import sklearn.metrics as metrics
y_predict=reg.predict(xtest)
```

```python
mse=metrics.mean_squared_error(ytest,y_predict)
mae=metrics.mean_absolute_error(ytest,y_predict)
rmse=np.sqrt(mse)
r2=metrics.r2_score(ytest,y_predict)
n=len(xtest)
adjusted_r2=((1-r2)*(n-1))/(n-1-1)
```

```python
print("MAE : ",mae)
print("MSE : ",mse)
print("RMSE : ",rmse)
print("R2 : ",r2)
print("Adjusted_R2 : ",adjusted_r2)
```
